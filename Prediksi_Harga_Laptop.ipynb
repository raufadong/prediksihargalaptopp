Masukkan Library yang digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Memanggil Dataset

```python
data = pd.read_csv('kliknklik_gaming_laptop.csv')
data.head()
```

Deskripsi Dataset

```python
data.info()
```

Normalisasi Dataset

```python
#case folding
import re
def casefolding(text):
  text = text.str.lower()
  return text
```

```python
#menghilangkan tanda baca
def number(nmbr):
  nmbr = re.sub(r'[^\w\s]', '', nmbr)
  return nmbr
```

```python
#sebelum dan setelah casefolding
raw_sample = data.iloc[:5]
case_folding = data.apply(casefolding)
Harga = case_folding['Harga'].apply(number)


print(raw_sample)
print(case_folding)
print(Harga)
```

```python
data = case_folding
data.head()
```

```python
data = data.drop(columns = ['Harga'])
data.head()
```

```python
data['Harga'] = Harga
data.head()
```

```python
#Normalisasi Kolom yang akan Digunakan Sebagai Fitur
data['RAM'] = data['RAM'].str.replace("gb",'')
data['RAM'] = data['RAM'].astype('int64')
data.rename(columns = {'RAM':'RAM(GB)'}, inplace = True)
data.head()
```

```python
data['Ukuran Layar'] = data['Ukuran Layar'].str.replace("inch",'')
data['Ukuran Layar'] = data['Ukuran Layar'].astype('int64')
data.rename(columns = {'Ukuran Layar':'Ukuran Layar(Inch)'}, inplace = True)
data.head()
```

```python
data['Harga'] = data['Harga'].str.replace("rp",'')
data['Harga'] = data['Harga'].astype('int64')
data.rename(columns = {'Harga':'Harga(RP)'}, inplace = True)
data.head()
```

```python
mc = data['Penyimpanan']
mct = mc.drop_duplicates().reset_index(drop=True)
print(mct)
```

```python
data['Penyimpanan'] = data['Penyimpanan'].str.replace("tb",'')
data['Penyimpanan'] = data['Penyimpanan'].str.replace("gb",'')
data['Penyimpanan'] = data['Penyimpanan'].astype('int64')
data.rename(columns = {'Penyimpanan':'Penyimpanan(GB)'}, inplace = True)
data.head()
```

```python
data_gb1 = data[data['Penyimpanan(GB)'] < 5]
data_gb2 = data[data['Penyimpanan(GB)'] > 5]
data_simpan1 = data_gb1['Penyimpanan(GB)']
data_simpan1 = data_simpan1 * 1000
data_simpan2 = data_gb2['Penyimpanan(GB)']
data_simpan1 = pd.DataFrame(data_simpan1)
data_simpan2 = pd.DataFrame(data_simpan2)
data_simpan = pd.concat([data_simpan1, data_simpan2])
print(data_simpan1)
print(data_simpan2)
print(data_simpan)
```

```python
data = data.drop(columns = ['Penyimpanan(GB)'])
data.head()
```

Visualisasi Data

```python
data['Penyimpanan(GB)'] = data_simpan
data.head()
```

```python
s = data['RAM(GB)'].value_counts()
plt.figure(figsize=(10,6))

s.plot(kind='bar')
plt.ylabel('Total')
plt.title('Laptop Berdasarkan Ukuran RAM')
plt.show()
```

```python
sns.barplot(x=data['RAM(GB)'],y=data['Harga(RP)'])
plt.show()
```

```python
b = data['Brand'].value_counts()
plt.figure(figsize=(10,6))

b.plot(kind='bar')
plt.ylabel('Total')
plt.title('Laptop Berdasarkan Brand')
plt.show()
```

Seleksi Fitur

```python
features = ['RAM(GB)', 'Ukuran Layar(Inch)', 'Penyimpanan(GB)']
x = data[features]
y = data['Harga(RP)']
x.shape, y.shape
```

split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,random_state=70)
y_test.shape
```

```python
#Membuat Model Regresi Linear
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
pred = lr.predict(x_test)
```

```python
score = lr.score(x_test, y_test)
print('akurasi model regresi linear', score)
```

Membuat Input Model Regresi Linear

```python
input_data = np.array([[8, 16, 1000]])

prediction = lr.predict(input_data)
print('estimasi harga laptop dalam Rp :', prediction)
```

Saving Model

```python
import pickle

filename = 'estimasi_laptop.sav'
pickle.dump(lr,open(filename,'wb'))
```
